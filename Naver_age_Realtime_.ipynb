```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
header = {'User-agent' : 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}  #헤더인식 
request = requests.get('https://datalab.naver.com/keyword/realtimeList.naver', headers = header)
print(request)

html = request.text
print(html)
```

```python
soup = BeautifulSoup(html,'html.parser')
outs = soup.findAll('div',{"data-age": "10s"})   # <class 'bs4.element.ResultSet'> => findAll() 메소드를 사용할 수 없다.
print(outs)  
 
#outs에서 html만 뽑아오는 과정    
#outs[0]  성분 하나 뽑아냄    # print(outs) -> [  outs[0]   ]  (값 1개인 리스트 )

ranks = outs[0].findAll('span',{"class":"title"})    # <class 'bs4.element.Tag'> => findAll() 메소드를 사용할 수 있다.    
print(ranks)
print(ranks[0])

for i in range(len(ranks)) :   #len[ ] x len() o 
    print("{}위 => {}".format(i+1,ranks[i].text))
```

데이터랩에서 연령대별로 파싱되는 범위가 달라지므로
딕셔너리를 이용하여(값입력) 파싱범위 조절 

```python
soup = BeautifulSoup(html,'html.parser')

#ranks = soup.findAll('span',{'class':'title'})
#labs = soup.findAll('div',{'class':'keyword_rank'})
        
while True :
    a = dict()
    a = { 0 : "all", 1: "10s", 2: "20s", 3 : "30s", 4 : "40s", 5 : "50s"}
    n = int(input("원하는 연령대의 정보를 입력> 0 : 전체, 1 : 10대, 2 : 20대, 3 : 30대, 4 : 40대, 5 : 50대 이상)"))
    
    if n > 6 :
        print("올바른 값을 입력하세요")
    else :
        k = a.get(n)
        break

outs = soup.findAll('div',{'data-age':k})
#outs = soup.findAll('div',{'data-age':'10s'})    # class 부분이 다른걸로 바뀌어서 뽑는 경우 


ranks = outs[0].findAll('span',{'class':'title'})   
   
for i in range(len(ranks)) :
    print( "{}위 => {}". format(i+1, ranks[i].text))
```

```python
# 딕셔너리 대신 유사 switch 문 

while True :
    n = int(input("원하는 연령대의 정보를 입력> 0 : 전체, 1 : 10대, 2 : 20대, 3 : 30대, 4 : 40대, 5 : 50대 이상)"))
    if n ==0 :
        k = 'all'
        break
    elif n == 1:
        k = '10s'
        break
    elif n == 2:
        k = '20s'
        break
    elif n == 3 :
        k = '30s'
        break
    elif n == 4 :
        k = '40s'
        break
    elif n == 5 :
        k = '50s'
        break
    else :
        print("올바른 값을 입력하세요")
```
