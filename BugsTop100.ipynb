```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
request = requests.get("https://music.bugs.co.kr/chart")
#print(request) #200

html = request.text
#print(html)

soup = BeautifulSoup(html,'html.parser')


musics = soup.findAll("p",{'class':'title'})
artists = soup.findAll("p",{'class':'artist'})


#print(musics[1].text.strip())


#print(type(artists[5].text.strip().split('\n'))) #list
#print(artists[5].text.strip().split('\n')[0] + artists[5].text.strip().split('\n')[3] )


a = list()
b = list()

now = dt.now()
print("현재 {}, 벅스뮤직의 TOP 200순위는\n\n".format(now))
for i in range(len(musics)):
    music = musics[i].text.strip()
    #artist = artists[i].text.strip().split('\n')
    
    #if len(artist) > 2 : 
    #   if artist[0] != artist[3] :
    #       artist = artist[0] + artist[3]
    #    else :
    #        artist = artist[0]
    #else :
    #     artist = artist[0]
            
    #통째로 
    artist = artists[i].text.strip().split('\n')[0]
    a.append(music)
    b.append(artist)
    
    print("{}위 - {} -{}".format(i+1,music,artist))

        
        
```

# 텍스트 파일에 저장된 크롤링된 데이터를 화면에 출력한다.
텍스트 파일을 open해서 데이터를 저장하려 할 때 UnicodeEncodeError: 'cp949' codec can't encode character가 발생되면 아래와 같이 파일을 열어서 UTF-8(유니코드)로 인코딩 시켜 저장하면 된다.

```python
file = open("BugsChart.txt",'w',-1,'UTF-8')

file.write("현재 {}\n\n".format(now))
for i in range(len(musics)) :
    data = "{}위 => {}-{}\n".format(i+1,a[i],b[i])
    file.write(data)
    
file.close()
print("등록완료")
```

```python
file = open("BugsChart.txt",'r',-1,'UTF-8')

lines = file.readlines()

for line in lines :
    if len(line.strip())!=0 :         # 공백까지 리스트안에 성분으로 들어가기 떄문 
        print(line,end=" ")          #띄어쓰기 제거 

file.close()
```
