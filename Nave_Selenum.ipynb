##Selenium에서 실행할 가상 크롬 웹 브라우저의 경로를 설정한다. 

```python
driver = webdriver.Chrome('./chromedriver.exe')  #C-대문자주의
```

#Selenium에서 가상 크롬을 통해 설정된 url의 사이트에 접속한다. 

```python
driver.get(targetSite)  #가상크롬에서 창이 뜸 
print(driver)
```

#Selenium으로 실행한 가상 크롬 브라우저의 페이지 소스를 얻어온다.

```python
html = driver.page_source
print(html)             #header 붙일 필요 없음 ,설정안해도됨 (가상 크롬 )  (익스플로러 x) 

soup = BeautifulSoup(html,'html.parser')
print(soup)
```

```python
ranks = soup.findAll('span',{'class':'title'})
#labs = soup.findAll('div',{'class':'keyword_rank'})
        
while True :
    a = dict()
    a = { 0 : "all", 1: "10s", 2: "20s", 3 : "30s", 4 : "40s", 5 : "50s"}
    n = int(input("원하는 연령대의 정보를 입력> 0 : 전체, 1 : 10대, 2 : 20대, 3 : 30대, 4 : 40대, 5 : 50대 이상)")) 
    if n > 6 :
        print("올바른 값을 입력하세요")
    else :
        k = a.get(n)
        break

outs = soup.findAll('div',{'data-age':k})
#outs = soup.findAll('div',{'data-age':'10s'})    # class 부분이 다른걸로 바뀌어서 뽑는 경우 
#labs = ranks.findAll('span',{'class':'title'})   #findAll() 전체 이용 불가능(값이 아닌 리스트)


ranks = outs[0].findAll('span',{'class':'title'})   
   
for i in range(len(ranks)) :
    print( "{}위 => {}". format(i+1, ranks[i].text))
```
